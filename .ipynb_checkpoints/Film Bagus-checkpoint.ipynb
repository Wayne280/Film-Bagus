```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.linear_model import LinearRegression
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
dmov = pd.read_csv('movies.csv')
dmov1 = dmov.copy()
drat = pd.read_csv('ratings.csv')
drat1 = drat.copy()
```

```python
##### Merger tabel movie dan rating #####
df = pd.merge(dmov1, drat1,how='inner', on='movieId')
df
```

```python
###### SOAL Pertama : Joko ######
```

```python
##### filter genre comdey|drama, sesuai selera Widodo #####
aa = df.loc[(df["genres"] == 'Action|Animation')]
aa
```

```python
##### cek movie favorit dari joko sebagai pembanding #####
joko = df.loc[(df["title"] == 'Superman vs. The Elite (2012)')]
joko
##### ratingnya hanya 3.0, kita tidak bisa mengunakan userID yang sama untuk dijadikan pembanding
```

```python
##### Rata-ratakan setiap rating dari movie yang sama #####
aaavg = aa.groupby('title').mean()
aaavg
#### TERNYATA tepat ada 5 movie selain Superman vs. The Elite (2012)
```

```python
##### Sortir bedasarkan rata-rata rating tertinggi #####
aarank = aaavg.sort_values(by='rating', ascending=False)
aarank
```
